```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
```

```python
df_tesla.head()
```

### Simple moving average



```python
df_tesla['Open'].plot(figsize=(15,6))
```

```python
# to create simple moving avg : from 18:00 min in video

df_tesla['Open:10days rolling'] = df_tesla['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:10days rolling']]
```

```python
df_tesla[['Open','Open:10days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```

##### in the above graph, we can see that orange is smoothening the curve


```python
# to create simple moving avg for 30days, 50 days etc

df_tesla['Open:30days rolling'] = df_tesla['Open'].rolling(window=30,min_periods=1).mean()
df_tesla['Open:40days rolling'] = df_tesla['Open'].rolling(window=40,min_periods=1).mean()
df_tesla['Open:50days rolling'] = df_tesla['Open'].rolling(window=50,min_periods=1).mean()
```

##### Still smoothened with 30 days and 50 days rolling

```python
df_tesla[['Open','Open:10days rolling','Open:30days rolling','Open:40days rolling','Open:50days rolling']].plot(xlim=['2020-01-01','2021-01-01'],figsize=(15,6))
```
