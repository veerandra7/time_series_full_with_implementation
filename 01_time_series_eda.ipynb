```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
```

```python
type(df_tesla)
```

```python
df_tesla.head()
```

```python
df_tesla.plot() # plots everything
```

```python
df_tesla['High'].plot(figsize=(12,4)) #plots one column
```

```python
## x limit and y limit

df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],figsize=(12,4))
```

```python
## y limit

df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],ylim=[0,900],figsize=(12,4))

```

```python
## y limit and x limit 

df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],ylim=[0,900],figsize=(12,4),c='green')

```

```python
## line style

df_tesla['High'].plot(xlim=['2020-01-01','2021-09-01'],ylim=[0,900],figsize=(12,4),c='green',ls='--')

```
